```python
# with open - create the inverted index. 
with open("Jep.csv","r") as f:
    reader =  csv.reader(f, delimiter=",")
    next(reader)
    for i, row in enumerate(reader):
        text = " ".join([row[3], row[5], row[6]]).split()
        text = [word.lower() for word in text]
        text = list(set(text))
        for word in text:
            if word not in inverted_idx:
                inverted_idx[word] = []
            inverted_idx[word].append(i)
print(inverted_idx)   
```

```python
""" Entity Recognition """
kind = []

for entity in nlp_string.ents:
    kind.append(entity.label_)
kind_count = Counter(kind)
top_kind = kind_count.most_common(5)
print(top_kind)

""" Sentence Count """
sentences = []
for token in nlp_string.sents:
    sentences.append(token)
print(len(sentences))    
```

```python
""" Nouns """
nouns = []
for token in nlp_string:
    if not token.is_stop | token.is_punct and token.pos_ == "NOUN":
        nouns.append(token.text)

        
"""" VERBS """
verbs = []
for token in nlp_string:
    if not token.is_stop | token.is_punct and token.pos_ == "VERB":
        verbs.append(token.text)
```
